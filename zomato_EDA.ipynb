```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
plt.pie(x=country_val,labels=country_names)
```

```python
# top 3 country
plt.pie(x=country_val[:3],labels=country_names[:3],autopct="%1.2f%%")
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating-count'})
```

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating-count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating-count',data=ratings,hue='Rating color')
```

```python
sns.barplot(x='Aggregate rating',y='Rating-count',data=ratings,hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```

```python
# count plot
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green'])
```
